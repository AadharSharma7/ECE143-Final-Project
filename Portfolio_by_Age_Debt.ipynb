```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_excel('Portfolio_by_Age_Debt_size.xls')
df
```

```python
age_df = df.drop(df.index[-3:])
age_df = age_df.iloc[6:]
age_df.drop(age_df.columns[[2, 4, 6, 8, 10]], axis=1, inplace=True)
age_df.columns = ['Debt Size', '24 and Younger', '25 to 34', '35 to 49', '50 to 61', '62 and Older']
age_df
```

```python
age_df = age_df.set_index(age_df['Debt Size'])
age_df.drop('Debt Size', axis=1, inplace=True)
age_df
```

```python
for col in age_df.columns:
    age_df[col] = age_df[col].astype(float)
s = sns.heatmap(age_df, annot=True, fmt=".1f")        #Dollars outstanding
plt.title("Debt size vs Age group Heatmap for Dollars outstanding")
plt.xlabel("Age group")
plt.ylabel("Debt size")
```

```python
age_df = df.drop(df.index[-3:])
age_df = age_df.iloc[6:]
age_df.drop(age_df.columns[[1, 3, 5, 7, 9]], axis=1, inplace=True)
age_df.columns = ['Debt Size', '24 and Younger', '25 to 34', '35 to 49', '50 to 61', '62 and Older']
age_df
```

```python
age_df = age_df.set_index(age_df['Debt Size'])
age_df.drop('Debt Size', axis=1, inplace=True)
age_df
```

```python
for col in age_df.columns:
    age_df[col] = age_df[col].astype(float)
sns.heatmap(age_df, annot=True, fmt=".1f")           #Number of borrowers
plt.title("Debt size vs Age group Heatmap for Number of borrowers")
plt.xlabel("Age group")
plt.ylabel("Debt size")
```

```python
plt.figure(figsize=(10, 6))

# Create line plot
sns.lineplot(data=age_df.T)  # Transpose the DataFrame for proper x-axis labeling

# Add title and axis labels
plt.title("Borrowers across age groups and for different debt sizes")
plt.xlabel("Age group")
plt.ylabel("Borrowers(in millions)")

# Rotate x-axis labels (optional)
plt.xticks(rotation=45, fontsize=8)

# Display the plot
plt.show()
```
